# 2. Train a CTGAN

```python
from sdv.single_table import CTGANSynthesizer
import pandas as pd
import json
```

## 2.1 Train a CTGAN with default Hyperparameters

```python
synthesizer_default = CTGANSynthesizer(
    metadata, 
    enforce_min_max_values = True, 
    enforce_rounding = True,
    epochs = 300,
    batch_size = 500, 
    generator_dim = (256, 256),
    discriminator_dim = (256,256), 
    verbose = True)
```

```python
%%capture output 

synthesizer_default.fit(data)
```

```python
# save model 
synthesizer_default.save(
    filepath = 'C:/Daten/DAV Ausbildung/AG/saved_models/ctgan_default.pkl'
)
```

```python
# sample synthetic data 

synthetic_data_default = synthesizer_default.sample(len(data.index))
#synthetic_data.head()
synthetic_data_default.describe()
```

```python
# write csv file
# for synthetic data 
synthetic_data_default.to_csv("C:/Daten/DAV Ausbildung/AG/saved_models/ctgan_default.csv")
```
